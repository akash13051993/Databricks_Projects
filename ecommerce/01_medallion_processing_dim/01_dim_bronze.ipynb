```python
import pyspark.sql.functions as F
catalog_name = 'ecommerce'
```

### Brands Table

```python
raw_data_path = '/Volumes/ecommerce/source_data/raw/brands/*.csv'

df = spark.read.option('header',"true") \
    .option('inferschema',"true") \
    .option('delimeter',",") \
    .csv(raw_data_path)

df = df.withColumn("_source_file" , F.col("_metadata.file_path"))\
       .withColumn("ingeted_at", F.current_timestamp())
       
display(df.limit(5))
```

```python
df.write.format("delta") \
        . mode("overwrite") \
            .option("mergeSchema", "true") \
                .saveAsTable(f"{catalog_name}.bronze.brz_brands")
```

### Catogory Table

```python
raw_data_path_category = '/Volumes/ecommerce/source_data/raw/category/*.csv'

df = spark.read.option('header',"true") \
            .option('inferSchema', "true") \
            .option('delimeter' , "true")\
            .csv(raw_data_path_category)

df = df.withColumn("_source_file", F.col("_metadata.file_path"))\
       .withColumn("ingested_at" , F.current_timestamp())

display(df.limit(5))
```

```python
df.write.format("delta") \
        .mode("overwrite")\
        .option("mergeSchema","true")\
            .saveAsTable(f"{catalog_name}.bronze.brz_category")
```
